### Flow control

* `if` , `else` , `elif`

* `while`, `break`, `continue`, `pass`

* `for` loop

```python
number = -102

if number > 0:
    print("Positive number")

elif number == 0:
    print('Zero')
else:
    print('Negative number')
```

```python
# program to display numbers from 1 to 5

# initialize the variable
i = 1
n = 5

# while loop from i = 1 to 5
while i <= n:
    print(i)
    i = i + 1
```

```python
# program to find first 5 multiples of 6

i = 1

while i <= 10:
    print('6 * ',(i), '=',6 * i)

    if i >= 5:
        break
    
    i = i + 1
```

```python
for i in range(5):
    if i == 3:
        continue
    print(i)
```

```python
n = 10

# use pass inside if statement
if n > 10:
    pass

print('Hello')
```

```python
# use of range() to define a range of values
values = range(4)

# iterate from i = 0 to i = 3
for i in values:
    print(i)
```
